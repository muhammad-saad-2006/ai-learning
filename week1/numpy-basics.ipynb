# Some NumPy Functions:

## Creating Numpy Arrays:

1. np.array([1,2,3,4]) <--- 1d array
2. np.array([[1,2,3], [4,5,6]]) <--- 2d array
3. We can use dtype to change data type like np.array([1,2,3], dtype=float)
4. np.arange(start, stop, step) i.e np.arange(1, 11, 2) creates an array like [1,3,5,7,9]
5. We can also use reshape to change the shape of created array like:<br>
np.arange(10).reshape(2,5) this will first create a 1d array with 10 elements and then convert it into matrix of 2x5. code example:

````python
import numpy as np

arr = np.arange(10).reshape(2, 5)
arr
````

6. np.ones((3,4)) <--- it creates a array of 3x4 size with only ones
7. np.zeros((3,4)) <--- same as above but creates matix having zero only
8. np.random.random((3,4)) <--- creates a array of given size with random values
9. np.linspace(start, end, number of elements in array) it is used for creating an array having values with equal distance from eachother. code example:

````python
arr1 = np.linspace(-10, 10, 10)
arr1
````

By default it has float vlaues to convert it into int just used a new parameter dtype like:

````python
arr1 = np.linspace(-10, 10, 10, dtype=int)
arr1
````

## Array Operation:

### 1. Scalar Operations
Scalar operations involve a vector and a single numeric value. The scalar value is applied to every single element inside the vector individually.

* **Scalar Addition:** Adds the scalar value to every element.
  ```python
  v + 5
  ```
* **Scalar Subtraction:** Subtracts the scalar value from every element.
  ```python
  v - 2
  ```
* **Scalar Multiplication:** Multiplies every element by the scalar value.
  ```python
  v * 3
  ```
* **Scalar Division:** Divides every element by the scalar value.
  ```python
  v / 2
  ```

### 2. Vector-to-Vector Operations
These operations involve two or more vectors. The math is performed element-by-element between corresponding positions, or it aggregates the vectors into a geometric result.

* **Vector Addition:** Adds corresponding elements of two equal-length vectors.
  ```python
  a + b
  ```
* **Vector Subtraction:** Subtracts corresponding elements of two equal-length vectors.
  ```python
  a - b
  ```
* **Element-wise Multiplication (Hadamard Product):** Multiplies corresponding elements together.
  ```python
  a * b
  ```
* **Element-wise Division:** Divides corresponding elements of the first vector by the second.
  ```python
  a / b
  ```
* **Dot Product:** Multiplies corresponding elements and returns the sum of those products as a single scalar.
  ```python
  np.dot(a, b)  # Or: a @ b
  ```
* **Cross Product:** Computes a new vector perpendicular to two vectors (primarily used in 3D space).
  ```python
  np.cross(a, b)
  ```
* **Vector Norm (Magnitude):** Calculates the spatial length or Euclidean distance of the vector.
  ```python
  np.linalg.norm(a)
  ```
