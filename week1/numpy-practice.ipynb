## Question 9:
Create a Python function to calculate the Softmax of the given numpy 1D array. The function only accepts the numpy 1D array, otherwise raise error.

$$\sigma(\vec{z})_i = \frac{e^{z_i}}{\sum_{j=i}^{K} e^{z_j}}$$


https://en.wikipedia.org/wiki/Softmax_function

**Example 1:**

Input:
```bash
[86.03331084 37.7285648  48.64908087 87.16563062 38.40852563 37.20006318]
```

Output:
```bash
[2.43733249e-01, 2.56112115e-22, 1.41628284e-17, 7.56266751e-01, 
       5.05514197e-22, 1.50974911e-22]
```

**Example 2:**

Input:
```bash
[33.17344305 45.61961654 82.05405781 80.9647098  68.82830233 91.52064278]
```

Output:
```bash
[4.57181035e-26, 1.16249923e-20, 7.73872596e-05, 2.60358426e-05,
       1.39571531e-10, 9.99896577e-01]
```

````python
def calculate_softmax(arr):
    if not isinstance(arr, np.ndarray):
        raise ValueError("Input must be a numpy array.")
    
    if arr.ndim != 1:
        raise ValueError("Input array must be one-dimensional.")
    
    safe_arr = arr - np.max(arr)

    exp_arr = np.exp(safe_arr)

    return exp_arr / np.sum(exp_arr)

input_array = np.array([86.03331084, 37.7285648, 48.64908087, 87.16563062, 38.40852563, 37.20006318])

output_array = calculate_softmax(input_array)
print(np.sum(output_array))  # This should print 1.0
print(output_array)
````

## Question 11:
Create a python function named **date_array** that accepts two dates as string format and returns a numpy array of dates between those 2 dates. The function only accept 2 strings, otherwise raise error. The date format should be like this only: `2022-12-6`. The end date should be included and for simplicity, choose dates from a same year.

**Example 1:**

Input:
```bash
date_array(start = '2020-09-15', end = '2020-09-25')
```

Output:
```bash
['2020-09-15', '2020-09-16', '2020-09-17', '2020-09-18',
 '2020-09-19', '2020-09-20', '2020-09-21', '2020-09-22',
 '2020-09-23', '2020-09-24', '2020-09-25']
```

**Example 2:**

Input:
```bash
date_array(start = '2022-12-01', end = '2022-12-06')
```

Output:
```bash
['2022-12-01', '2022-12-02', '2022-12-03', '2022-12-04', '2022-12-05', '2022-12-06']
```

**Example 3:**

Input:
```bash
date_array(start = '2020-11-25', end = '2020-11-30')
```

Output:
```bash
['2020-11-25', '2020-11-26', '2020-11-27', '2020-11-28',
 '2020-11-29', '2020-11-30']
```

````python
def date_array(start_date, end_date):
    start = np.datetime64(start_date)
    end = np.datetime64(end_date) + np.timedelta64(1, 'D') 
    return np.arange(start, end)

print(date_array('2022-12-01','2022-12-06'))
````

## Question 14:
Answer below asked questions on given array:
1. Fetch Every alternate column of the array
2. Normalise the given array

https://en.wikipedia.org/wiki/Normalization_(statistics)

There are different form of normalisation for this question use below formula.

$$\large X_{normalized} = \frac{X - X_{min}}{X_{max} - X_{min}}$$

```python
arr1=np.random.randint(low=1, high=10000, size=40).reshape(8,5)
```


````python
arr = np.random.randint(low=1, high=10000, size=40).reshape(8, 5)

# 1. fetching alternative columns
alt_cols = arr[:, ::2]
print("Alternative columns:\n", alt_cols)

# 2. Normalizing the given array
arr_min = np.min(arr)
arr_max = np.max(arr)
normalized_arr = (arr-arr_min) / (arr_max - arr_min)
print("Normalized array:\n", normalized_arr)
````

## Question 15:
Create the following pattern without hardcoding. Use only numpy functions and the below input array a.
```
# Input: a = np.array([1,2,3])
# Output: array([1, 1, 1, 2, 2, 2, 3, 3, 3, 1, 2, 3, 1, 2, 3, 1, 2, 3])
```

````python
a = np.array([1, 2, 3 ])

part1 = np.repeat(a, 3) # repeat is used to repat each element of the array a specified number of times.
part2 = np.tile(a, 3) # tile is used to repeat the whole array a specified number of times.

final_array = np.concatenate([part1, part2])
print("Final array:\n", final_array)
````

## Question 19:
Your are given an array which is havig some nan value. You job is to fill those nan values with most common element in the array.
```
arr=np.array([[1,2,np.nan],[4,2,6],[np.nan,np.nan,5]])

```

````python
arr = np.array([[1, 2, np.nan], [4, 2, 6], [np.nan, np.nan, 5]])

values, count = np.unique(arr[~np.isnan(arr)], return_counts=True)

most_common = values[np.argmax(count)]

arr[np.isnan(arr)] = most_common

print(arr)
````

## Question 22:
Plot this below equation.

$$\large y = \frac{e^x - e^{-x}}{e^x + e^{-x}}$$

**Note: This equation is called tanh activation function. In deep learning, many times this function is used.

````python
import matplotlib.pyplot as plt

x = np.linspace(10, -10, 100)

y = (np.exp(x) - np.exp(-x)) / (np.exp(x) + np.exp(-x))

plt.plot(x, y);
````
