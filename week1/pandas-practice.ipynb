## Question 3:
Write a function to change the data type of given a column or a Series. Function takes series and data type as input, returns the converted series.
```
series = pd.Series([1,2,'Python', 2.0, True, 100])
change to float type data
```
Note: Read about pd.to_numeric()

````python
def convert_type(series, dtype):
    
    if dtype in ['int', 'float', 'float64', 'int64']:
        return pd.to_numeric(series, errors='coerce')
    else:
        return series.astype(dtype)

series = pd.Series([1, 2, 'Python', 2.0, True, 100])
converted_series = convert_type(series, 'float')

print(converted_series)
````

````python
def dashboard(t1, t2):
   # first find both teams data
   teams_data = df[
      ((df['Team1'] == t1) | (df['Team2'] == t1)) &
      ((df['Team1'] == t2) | (df['Team2'] == t2)) 
   ]
   display(teams_data)
   print(50*"-")
   print("\n")

   # find total number of matches played
   print(15*"-", "Total matches played", 15*"-")
   no_of_matches_played = teams_data['Team1'].value_counts().add(teams_data['Team2'].value_counts(), fill_value=0)
   total_matches_played = no_of_matches_played.reset_index()
   total_matches_played.columns = ['Teams Name', 'No. of matched played']
   print(total_matches_played)
   print(50*"-")
   print("\n")

   # find the win record of teams
   print(15*"-", "Win Record of teams", 15*"-")
   win_record = teams_data['WinningTeam'].value_counts()
   wins = win_record.reset_index()
   wins.columns = ['Teams Name', 'No. of matches won']
   print(wins)
   print(50*"-")
   print("\n")

   # find the loss record of teams
   print(15*"-", "Loss Record of teams", 15*"-")
   loss_record = no_of_matches_played.sub(win_record, fill_value=0)
   losses = loss_record.reset_index()
   losses.columns = ['Teams Name', 'No. of matches lost']
   print(losses)
   print(50*"-")
   print("\n")

   # now finding the most player of the matches award
   print(15*"-", "Player of Matches Record", 15*"-")
   player_of_match = teams_data['Player_of_Match'].value_counts().head(1)
   print(player_of_match)
   print(50*"-")


dashboard('Kolkata Knight Riders', 'Chennai Super Kings')
````

## Question 1:
Perform `groupby` based on the `"Team"` column and then perform Z Normalization on top of the below columns of each group:
1. Passes
2. Passes Completed
3. Attempted Line Breaks
4. Completed Line Breaks

You have to make a python function named `z_normalization` which takes two arguments:

1. *group:* Every group that you have created
2. *cols_to_perform:* This parameter takes a list of columns on which you have to perform the Z-Normalization.

For this problem, you have to use th `apply()` method.

$\Large Z \ - \ Normalization = \frac{X_i - \mu}{std} $

After that find out the below values for each group:
- minimum "Passess"
- maximum "Passess"
- minimum "Yellow Cards"
- maximum "Yellow cards"
- average "Yellow Cards"
- maximum "Attempted Line Breaks"
- minimum "Attempted Line Breaks"
- standardard deviation of "Attempted Line Breaks"
- average Possession

````python
def z_normalization(group, cols_to_perform):
    
    for col in cols_to_perform:

        mu = group[col].mean()

        std = group[col].std()

        if std == 0:
            group[col] = group[col] - mu
        else:
            group[col] = (group[col] - mu) /std
    return group

columns = ['Passes', 'Passes Completed', 'Attempted Line Breaks', 'Completed Line Breaks']
fifa_df[columns] = fifa_df[columns].apply(pd.to_numeric)
fifa_df[columns] = fifa_df.groupby("Team")[columns].apply(z_normalization, cols_to_perform=columns).reset_index(level=0, drop=True)


print(fifa_df.groupby("Team")['Passes'].agg(['min', 'max'])) 
print(fifa_df.groupby("Team")['Yellow Cards'].agg(['min', 'max', 'mean'])) 
print(fifa_df.groupby("Team")['Attempted Line Breaks'].agg(['max', 'min', 'std']))
print(fifa_df.groupby("Team")['Possession (%)'].agg(['mean']))  
````

## Queation1: 
Find batsman in below category-
* Highest score while chasing
* Best Strike rate while chasing and have faced 100+ balls


> Chasing mean team batting in second inning

````python
ipl_df = pd.read_csv("E:/ai-learning/data/ipl_deliveries.csv")

# first filter the table for 2nd innings only
chasing_team = ipl_df[ipl_df['innings'] == 2]

#exclude wide deliveries from ball count
valid_deliveries = chasing_team[(chasing_team['extra_type'].isna()) | (~chasing_team['extra_type'].isin(['wides', 'wide']))] 


# find the high scoring batsman
high_scoring_batsman = chasing_team.groupby('batter')['batsman_run'].sum().sort_values(ascending=False).head(1)
display(high_scoring_batsman)

# now find the batsman with best strike rate

#first of all separating batsman stats of ball faced and total runs
batsman_stats = valid_deliveries.groupby('batter').agg(
    Ball_faced=('ballnumber', 'count'),
    Total_runs= ('batsman_run', 'sum'))
display(batsman_stats)

# now filter the stats include only batsman who have faced 100 balls
balls_above_100 = batsman_stats[batsman_stats['Ball_faced'] >= 100]

# now find the batter with best strike rate
balls_above_100['Strike_Rate'] = (balls_above_100['Total_runs'] / balls_above_100['Ball_faced']) * 100
print(balls_above_100.sort_values(by='Strike_Rate', ascending=False).head(1))
````

## Question 2:
 Most Successful bowler against any batsman. Find that pair of bowler and batsman.
> Most Successful in terms of dissmissal. A bowler who have dissmissed any batsman most no of times. If any two pairs have same no of dissmisal, consider runs conceded by bowler to that batsman. Those who have concede lesser runs is more successful.

````python
# first filter the data with rows having dismissal value is 1
dismissal_df = ipl_df[ipl_df['isWicketDelivery'] == 1]
#display(dismissal_df)

# pair bowler and batsman on dismissal
pair_stat = dismissal_df.groupby(['bowler', 'batter']).agg( dismissal= ('isWicketDelivery', 'count')).reset_index()
#print(pair_stat)

# add runs conceded
runs_df = ipl_df.groupby(['bowler', 'batter']).agg(runs_conceded= ('batsman_run', 'sum')).reset_index()
#print(runs_df)

# now merge both results
final_stats = pair_stat.merge(runs_df, on=['bowler', 'batter'])
#print(final_stats)

# now finding most successful bowler
result = final_stats.sort_values(by=['dismissal', 'runs_conceded'], ascending= (False, True))
print(result.head(1))
````

## Question 3:
 Most successful batting pair in IPL. Batting pair who have scored most runs playing together.

````python
ipl_df[['p1', 'p2']] = pd.DataFrame(
    ipl_df.apply(lambda x: sorted([x['batter'], x['non-striker']]), axis=1).tolist(),
    index=ipl_df.index
)

batting_pairs = ipl_df.groupby(['p1', 'p2']).agg(
    runs_scored = ('total_run', 'sum')
).reset_index()

print(batting_pairs.sort_values(by='runs_scored', ascending=False).head(1))
ipl_df.drop(columns=['p1', 'p2'], inplace=True)
````

## Question 4:
Make a dataframe for all batting pairs played together.
```
Batsman1 Batsman2 Runs Avg StrikeRate
```

> Just to ease this question you can count wide-balls for strike rate

````python
ipl_df[['Batsman1', 'Batsman2']] = pd.DataFrame(
    ipl_df.apply(lambda x: sorted([x['batter'], x['non-striker']]), axis=1).tolist(),
    index=ipl_df.index
)

result = ipl_df.groupby(['Batsman1', 'Batsman2']).agg(
    Runs = ('total_run', 'sum'),
    Balls = ('ballnumber', 'count'),
    Dismissals = ('isWicketDelivery', 'sum')
).reset_index()

result['Avg'] = result.apply(lambda x: x['Runs'] / x['Dismissals'] if x['Dismissals'] > 0 else x['Runs'], axis=1)
result['Strike_Rate'] = (result['Runs'] / result['Balls']) * 100
print(result.sort_values(by=['Avg'], ascending=False))
````
