## Question 1:
 Draw a line plot of which, the x-axis is the "Year" and the y-axis is sum of "PM2.5" of two countries Iran and China. 

```python
df = pd.read_csv("E:/ai-learning/data/List of most-polluted cities by particulate matter concentration - List of most-polluted cities by particulate matter concentration.csv")
#display(df.head())

countries = ['Iran', 'China']
country_df = df[df['Country'].isin(countries)]

data = country_df.groupby(['Year', 'Country'])['PM2.5'].sum().reset_index()

iran_df = data[data['Country'] == 'Iran']
china_df = data[data['Country'] == 'China']

plt.figure(figsize=(10,5))
plt.plot(iran_df['Year'], iran_df['PM2.5'], label='Iran', linewidth=3)
plt.plot(china_df['Year'], china_df['PM2.5'], label='China', linewidth=3)

plt.title("Year vs Sum of PM2.5 (Iran and China)")
plt.xlabel('Years')
plt.ylabel('Sum of PM2.5')
plt.xticks(data['Year'])
plt.grid()

plt.legend()
plt.show()
```

## Question 2:
Draw a histogram of the column "PM10" of which the y-axis represents the probability (see the documentation how to draw the probability).

```python
plt.hist(df['PM10'], density=True, bins=50, facecolor= 'purple')
plt.xlabel('Bins')
plt.ylabel('Pribability')
plt.title('Histogram of PM10')
plt.grid()

plt.show()
```

## Question 3:
Draw a scatter plot where x-axis represents "PM2.5" and y-axis represents "PM10" for two countries Poland and Chile

```python
poland_df = df[df['Country'] == 'Poland']
chile_df = df[df['Country'] == 'Chile']

plt.scatter(poland_df['PM2.5'], poland_df['PM10'], marker='+', color='red', linewidth=3, s=75, label='Poland')
plt.scatter(chile_df['PM2.5'], chile_df['PM10'], marker='o', color='green', label='Chile')
plt.xlabel('PM2.5')
plt.ylabel('PM10')
plt.title('PM2.5 vs PM10 for Chile and Poland')
plt.grid()
plt.legend()

plt.show()
```

## Question 4:
Draw a pie chart of top 5 most frequent countries.

```python
top_5_countries = df["Country"].value_counts().head(5)

plt.pie(top_5_countries, labels=top_5_countries.index, autopct='%0.1f%%', shadow=True)
plt.show()
```

## Question 5:
Draw a bar chart which represents the counts of top 5 most frequent countries.

```python
plt.bar(top_5_countries.index, top_5_countries)
plt.title("Top 5 frequent Countries")
plt.xlabel('Countries')
plt.ylabel('Frequency Count')
plt.show()
```
