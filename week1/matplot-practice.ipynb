## Question 1:
 Draw a line plot of which, the x-axis is the "Year" and the y-axis is sum of "PM2.5" of two countries Iran and China. 

```python
df = pd.read_csv("E:/ai-learning/data/List of most-polluted cities by particulate matter concentration - List of most-polluted cities by particulate matter concentration.csv")
#display(df.head())

countries = ['Iran', 'China']
country_df = df[df['Country'].isin(countries)]

data = country_df.groupby(['Year', 'Country'])['PM2.5'].sum().reset_index()

iran_df = data[data['Country'] == 'Iran']
china_df = data[data['Country'] == 'China']

plt.figure(figsize=(10,5))
plt.plot(iran_df['Year'], iran_df['PM2.5'], label='Iran', linewidth=3)
plt.plot(china_df['Year'], china_df['PM2.5'], label='China', linewidth=3)

plt.title("Year vs Sum of PM2.5 (Iran and China)")
plt.xlabel('Years')
plt.ylabel ('Sum of PM2.5')
plt.xticks(data['Year'])
plt.grid()

plt.legend()
plt.show()
```

### **Problem-1:** Make a subplots which have 2 plots.

- For the first chart, draw a scatter plot "*Monitored Cap.(MW)*" vs "*Total Cap. Under Maintenance (MW)*" of top 5 most frequent power stations. Then draw the lines which indicate the average values of these two columns. Change the colors according to the names of the Power Stations.
- For the second chart, draw a scatter plot "*Monitored Cap. (MW)*" vs "*Actual(MU)*" of the top 5 most frequent power stations. Also draw the lines which indicates the average values of these two columns. Change the colors according to the names of the Power Stations.


```python
df = pd.read_csv("E:/ai-learning/data/PowerGeneration - PowerGeneration.csv")

# we need to make a scatter plot of top 5 frequent power stations
frequent_power_stations = df['Power Station'].value_counts().head(5).reset_index()

frequent_power_stations = df[df['Power Station'].isin(frequent_power_stations['Power Station'])]

# now find the avg values of these col Monitored Cap.(MW) and Total Cap. Under Maintenace (MW)
avg_monitored = frequent_power_stations['Monitored Cap.(MW)'].mean()
avg_maintenance = frequent_power_stations['Total Cap. Under Maintenace (MW)'].mean() 
avg_actual_mu = frequent_power_stations['Actual(MU)'].mean()

# graph 1:
fig, ax = plt.subplots(nrows=2, ncols=1, sharex=True, figsize=(12,8))

for station, station_data in frequent_power_stations.groupby('Power Station'):
    ax[0].scatter(station_data['Monitored Cap.(MW)'], station_data['Total Cap. Under Maintenace (MW)'], label=station)

ax[0].set_ylabel('Total Cap. Under Maintenace (MW)')
ax[0].set_title('Monitored vs Maintenance (Top 5 frequent Power Stations)')
ax[0].axvline(avg_monitored, color='green', linestyle='dashed', label='Average Monitored')
ax[0].axhline(avg_maintenance, color='red', linestyle='dashed',  label='Average Maintenance')
ax[0].grid()
ax[0].legend() 

# graph 2
for station, station_data in frequent_power_stations.groupby('Power Station'):
    ax[1].scatter(station_data['Monitored Cap.(MW)'], station_data['Actual(MU)'], label=station)
ax[1].set_xlabel('Monitored Cap.(MW)')
ax[1].set_ylabel('Actual(MU)')
ax[1].axvline(avg_monitored, color='green', linestyle='dashed', label='Average Monitored')
ax[1].axhline(avg_actual_mu, color='red', linestyle='dashed',  label='Average Actual(MU)')
ax[1].grid()
ax[1].legend()

plt.show()
```

## **Problem-2:** 
Draw a 3D Scatter plot between "*Monitored Cap.(MW)*", "*Total Cap. Under Maintenace (MW)*" and "*Forced Maintanence(MW)*"

```python
fig = plt.figure(figsize=(12,8))
ax = plt.subplot(projection='3d')

ax.scatter3D(df['Monitored Cap.(MW)'], df['Total Cap. Under Maintenace (MW)'], df['Forced Maintanence(MW)'])
ax.set_xlabel('Monitored Cap.(MW)')
ax.set_ylabel('Total Cap. Under Maintenace (MW)')
ax.set_zlabel('Forced Maintanence(MW)', labelpad=-1)

plt.show()
```
