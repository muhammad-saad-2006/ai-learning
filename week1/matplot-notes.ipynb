## `3d Plot:`

### `3d Scatter Plots:`

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

batter_df = pd.read_csv("E:/ai-learning/data/batter.csv")

fig = plt.figure(figsize=(12,8))
ax = plt.subplot(projection='3d') 
ax.scatter3D(batter_df['runs'], batter_df['avg'], batter_df['strike_rate']) #----> gets 3 axis x, y, z
ax.set_title('IPl batters data on 3d Scatter plot')
ax.set_xlabel('Runs')
ax.set_ylabel('Average')
ax.set_zlabel('Strike Rate', labelpad=-1)

plt.show()
```

## `3d line Plot:`

```python
x = [0,1,5,25]
y = [0,10,13,0]
z = [0,13,20,9]

fig = plt.figure(figsize=(10, 6))

ax = plt.subplot(projection='3d')

ax.scatter3D(x,y,z,s=[100,100,100,100])
ax.plot3D(x,y,z,color='green', linestyle='dashdot')

plt.show()
```

## `3d Surface Plot:`

```python

## plotting the graph of z = x^2 + y^2

x = np.linspace(-10, 10, 100)
y = np.linspace(-10, 10, 100)

xx, yy =np.meshgrid(x,y)

z = xx**2 + yy**2

fig = plt.figure(figsize=(12, 8))
ax = plt.subplot(projection='3d')

graph = ax.plot_surface(xx, yy, z, cmap='viridis')
fig.colorbar(graph)

plt.show()
```

## `Contour Plot:`

Contour is type of grpah that converts the 3d graphs into 2d graph. Lets take a walkthrough example taking the above example where we plotted:
> z = x^2 + y^2

```python
fig = plt.figure(figsize=(12, 8))
ax = plt.subplot() # didn't used 3d here because we need to plot a 2d graph

graph = ax.contour(xx, yy, z, cmap='viridis')
fig.colorbar(graph)

plt.show()
```

We can also fill the color by useing contourf to see color filled graph which help us to understand it visually.

```python
fig = plt.figure(figsize=(12, 8))
ax = plt.subplot() 

graph = ax.contourf(xx, yy, z, cmap='viridis')
fig.colorbar(graph)

plt.show()
```

## `Heatmap:`

```python
df = pd.read_csv("E:/ai-learning/data/IPL_Ball_by_Ball_2008_2022.csv")

df.head()
```

```python
temp_df = df[df['ballnumber'].isin([1, 2, 3, 4, 5, 6]) & (df['batsman_run'] == 6)] 
grid = temp_df.pivot_table(index='overs', columns='ballnumber', values='batsman_run', aggfunc='count') 
display(grid)            
```
