## `1. Relational Plot:`

Plots under this section

```python
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px # it contains toy data set which can be used for analysis
```

```python
tips = sns.load_dataset('tips')
tips
```

```python
# plotting a scatter plot using sns
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='sex', style='time', size='size') # first pass the dataset then your x and y axis as shown
```

In seaborn there are 2 types of functions:
- Figure Level
- Axis Level
in above example we have ploted the graph using axis level function we can plot the same graph using figure level function, it is as follow

```python
sns.relplot(data=tips, x='total_bill', y='tip', kind='scatter', hue='sex', style='time', size='size')
```

### Figure-Level Functions

- Figure-level functions create and manage the entire figure automatically.
- They are useful for creating multiple plots using facets such as rows and columns.
- These functions return a Seaborn `FacetGrid`, `PairGrid`, or `JointGrid` object.
- They provide a high-level interface with built-in support for complex visualizations.
- Examples include `relplot()`, `catplot()`, `displot()`, and `lmplot()`.

### Axis-Level Functions

- Axis-level functions draw plots on a specific Matplotlib Axes object.
- They offer more control when customizing individual plots.
- These functions are commonly used when working with Matplotlib subplots.
- They return a Matplotlib `Axes` object instead of creating a new figure.
- Examples include `scatterplot()`, `lineplot()`, `histplot()`, `boxplot()`, and `barplot()`.

```python
# Line plot using seaborn
gap = px.data.gapminder()
temp_df = gap[gap['country'] == 'Pakistan']
display(temp_df)
```

```python
sns.lineplot(temp_df, x='year', y='lifeExp')
```

```python
# hue and style parameter in line plot
temp = gap[gap['country'].isin(['Brazil', 'Pakistan', 'Germany'])]

sns.relplot(data=temp, x='year', y='lifeExp', kind='line', hue='country', style='continent')
```

```python
# there is plot called facet plot used to plot multiple graph in same figure side by side
# facet plot only works with figure level functions
sns.relplot(data=tips, x='total_bill', y='tip', kind='scatter', col='smoker', row='sex', hue='day') 
```

### KDE Plot (Kernel Density Estimation)

- A KDE plot is a smooth curve that estimates the probability density of a continuous variable.
- It is used to visualize the distribution of data without using histogram bars.
- The peaks of the curve represent regions where data points are more concentrated.
- KDE plots are useful for comparing the distributions of different categories using the `hue` parameter.
- Seaborn provides the `kdeplot()` function, and a KDE curve can also be added to a histogram using `kde=True`.

```python
sns.kdeplot(data=tips, x='total_bill', hue='sex', fill=True)
```

### Rug Plot

- A rug plot displays every individual data point as a small tick mark along an axis.
- It is useful for visualizing the exact distribution and concentration of observations.
- Closely spaced ticks indicate dense regions, while gaps indicate sparse regions.
- Rug plots are commonly combined with KDE or histogram plots for better interpretation.
- Seaborn provides the `rugplot()` function to create rug plots for one-dimensional data.

```python
sns.kdeplot(data=tips, x='total_bill')
sns.rugplot(data=tips, x='total_bill')
```

### Boxplot:

A boxplot is a standardized way of displaying the distribution of data based on a five number summary (“minimum”, first quartile [Q1], median, third quartile [Q3] and “maximum”). It can tell you about your outliers and what their values are. Boxplots can also tell you if your data is symmetrical, how tightly your data is grouped and if and how your data is skewed.

image.png

```python
sns.boxplot(data=tips, x='day', y='total_bill', hue='sex')
```

```python
sns.boxplot(data=tips, y='total_bill')
```

```python
# violinplot (is is made from boxplot+KDEplot)

sns.violinplot(data=tips, x='day', y='total_bill', hue='sex', split=True)
```

## Regression Plots:

- regplot
- lmplot

In the simplest invocation, both functions draw a scatterplot of two variables, x and y, and then fit the regression model y ~ x and plot the resulting regression line and a 95% confidence interval for that regression.

```python
# regplot is axis level function and it does not provide hue
sns.regplot(data=tips, x='total_bill', y='tip')
```

```python
# lmplot is figure level function of regression plot and it alos provide hue 
sns.lmplot(data=tips, x='total_bill', y='tip', hue='sex')
```

```python
# residplot
sns.residplot(data=tips, x='total_bill', y='tip')
```

### A second way to plot Facet plots --> FacetGrid

```python
sns.catplot(data=tips, x='sex', y='total_bill', kind='violin', col='day', row='time')
```

```python
g = sns.FacetGrid(data=tips,col='day',row='time')
g.map(sns.violinplot, 'sex', 'total_bill')

plt.show()
```

### Plotting Pairwise Relationship:

```python
sns.pairplot(iris, hue='species')
```

```python
# pair grid
g = sns.PairGrid(data=iris, hue='species')
g.map(sns.scatterplot)
```

```python
g = sns.PairGrid(data=iris, hue='species')
g.map_diag(sns.histplot)
g.map_offdiag(sns.scatterplot)
```

```python
g = sns.PairGrid(iris)
g.map_diag(sns.histplot)
g.map_upper(sns.kdeplot)
g.map_lower(sns.scatterplot)
```

```python
sns.jointplot(data=tips, x='total_bill', y='tip', kind='hex')
```

```python
g = sns.JointGrid(data=tips, x='total_bill', y='tip')
g.plot(sns.scatterplot, sns.histplot)
```
