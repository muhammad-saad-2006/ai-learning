## `What is Data Analysis:`

> Data analysis is the process of inspecting, cleansing, tranforming and modeling data with the goal of discovering usedful information, informing conclusion and supporting decision-making.

## `Data Analysis Process:`

Data is analyzed in 5 steps which are as follow:

1. **Asking Questions:** 
Asking right question like what is the data about why we need such data what we can do with this data.

2. **Data Preprocessing:**
The data we get is messy we can not use it for getting a meaningful result in order to achieve it we need to clean the data and then use it for analysis.

3. **Exploratory Data Analysis(EDA):**
Once we have cleaned our data we do exploration.  Analyzing the data through statistical summaries and visual graphs (histograms, scatter plots, etc.) to understand relationships.
Improving the data to facilitate better insights, which includes removing outliers and feature engineering (creating new columns or transforming existing ones).

4. **Drawing Conclusions:**
Using the insights gained, you draw conclusions or make predictions. This often involves Machine Learning algorithms or Inferential Statistics to make predictions about a larger population based on a sample.

5. **Communicating Results:**
The final step is effectively communicating your findings. It’s about translating technical analysis into actionable insights that can support business decision-making.

### `12. Converters:`

```python
def name_converter(team):
    words = team.split()
    short_name = ''

    for word in words:
        short_name += word[0].upper()
    return short_name
```

```python
pd.read_csv('E:/ai-learning/data/Data for week-2/IPL Matches 2008-2020.csv', converters=({'team1': name_converter, 'team2': name_converter}))
```

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
from urllib.parse import urljoin
import time

url = "https://books.toscrape.com/catalogue/page-1.html"
i = 1
books_data = []

while url:

    try:
        response = requests.get(url, timeout=10)
        response.raise_for_status()
    except requests.exceptions.RequestException as e:
        print(f'failed to open: {url}\n{e}' ) 
        break

    response.encoding = 'utf-8'

    html = response.text

    soup = BeautifulSoup(html, 'html.parser')

    books = soup.find_all('article', class_='product_pod')

    for book in books:
        title = book.h3.a['title']

        price = book.find('p', class_='price_color').text.replace('£', '')

        rating = book.find("p", class_="star-rating")["class"][1]

        availability = book.find('p', class_='instock availability').text.strip()

        image_container = book.find('div', class_='image_container')
        image_url = image_container.find('img')['src'] if image_container else 'No image'

        relative_link = book.h3.a['href']
        book_url = urljoin(url, relative_link)

        try:
            book_response = requests.get(book_url, timeout=10)
            book_response.encoding = 'utf-8'

            book_soup = BeautifulSoup(book_response.text, 'html.parser')

            category = book_soup.find('ul', class_='breadcrumb').find_all('li')[2].text.strip()
        except requests.exceptions.RequestException:
            category = None
        books_data.append({
            "title": title,
            "category": category,
            "price": float(price),
            "rating": rating,
            "availability": availability,
            "image_url": image_url
        })
        
        print(f'{i}. ',title, category, price, rating, availability, image_url)
        i += 1
    next_button = soup.find('li', class_='next')

    if next_button:
        next_page_link = next_button.find('a')['href']

        url = urljoin(url, next_page_link)
    else:
        url =None

        
```
