## `What is Data Analysis:`

> Data analysis is the process of inspecting, cleansing, tranforming and modeling data with the goal of discovering usedful information, informing conclusion and supporting decision-making.

## `Data Analysis Process:`

Data is analyzed in 5 steps which are as follow:

1. **Asking Questions:** 
Asking right question like what is the data about why we need such data what we can do with this data.

2. **Data Preprocessing:**
The data we get is messy we can not use it for getting a meaningful result in order to achieve it we need to clean the data and then use it for analysis.

3. **Exploratory Data Analysis(EDA):**
Once we have cleaned our data we do exploration.  Analyzing the data through statistical summaries and visual graphs (histograms, scatter plots, etc.) to understand relationships.
Improving the data to facilitate better insights, which includes removing outliers and feature engineering (creating new columns or transforming existing ones).

4. **Drawing Conclusions:**
Using the insights gained, you draw conclusions or make predictions. This often involves Machine Learning algorithms or Inferential Statistics to make predictions about a larger population based on a sample.

5. **Communicating Results:**
The final step is effectively communicating your findings. It’s about translating technical analysis into actionable insights that can support business decision-making.
